## (1) 주제 선정 이유

 파이썬 질문 게시판에서 많이 쓰였던 단어가 무엇인지 궁금하였다. 웹/파이선프로그래밍 강의를 듣고 질의 게시판을 이용하다 보니 중복되는 질문도 많이 보였다. 그렇다면 중복되는 단어들은 어떤 단어들이 있을지 궁금증이 생겨 워드 클라우드를 만들어 보기로 하였다. 시각화를 통해 어떤 단어들이 많이 쓰였는지 한 눈에 볼 수 있기에 궁금증을 해결하기 적합하다는 생각이 들었다. 인터넷을 통해 정보를 수집하는 것도 이에 적합하다는 생각이 들어 이 주제를 선정하게 되었다.


## (2) 가설 정의

 질문 게시판이다보니 '@@@ 질문 있습니다','### 질문' 등과 같은 글들로 '질문'이라는 단어가 많이 쓰였을 것이다. 그 외에 다음으로 많이 쓰이는 단어들을 통해 어떤 질문이 많이 나왔는지 추측할 수 있을 것이다. 예를 들면 'ipynb' 라는 단어가 워드클라우드에서 보인다면 파일 확장자 이슈에 대한 질문이 많았을 것으로 추정할 수 있다. 이러한 방법을 통해 FAQ(Frequently asked questions)가 모여있는 글을 만드는 등 좀더 효율적인 질문 게시판이 되는 데에 기여할 수 있을 것이다.

전처리 시작

```python
import re

# removing special characters

def clean_text(string): 
    rm_sp = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』_‘|\(\)\[\]`\'…》\”\“\’·0-9\r\n]', '', string)
    return rm_sp


# checking if its korean

def is_it_kor(word):
    if word.encode().isalpha():
        return False
    else:
        return True
    
# removing stopwords

def rem_stopwords(wordlist, stopwords):
    num = len(wordlist)
    for i in stopwords:
        for j in range(num):
            if i in wordlist[j]:
                temp = wordlist[j]
                wordlist[j] = temp.rstrip(i)
    return wordlist

# removing trash words

def rem_trash(data):
    result = [i for i in data if len(i) > 1] # 한 글자 이하의 데이터 삭제
    return result
```

```python
# raw data 불러오기

with open('raw_titles.csv', newline='',encoding='utf-8-sig') as f:
    titles_data = list(f)

with open('raw_contents.csv', newline='',encoding='utf-8-sig') as f:
    contents_data = list(f)
```

```python
for i in range(len(titles_data)):   # 만들어둔 특수문자 제거 함수 이용
    clean_temp = clean_text(titles_data[i])
    titles_data[i] = clean_temp
    
for i in range(len(contents_data)):
    clean_temp = clean_text(contents_data[i])
    contents_data[i] = clean_temp
    
    
if titles_data[0] == 'titles':
    titles_data.remove('titles') # 컬럼 이름 삭제

if contents_data[0] == 'contents':
    contents_data.remove('contents')
```

```python
# 한글과 영어 리스트 나누기

kor_titles = []
eng_titles = []

for i in titles_data: 
    if is_it_kor(i) == True and len(i) <= 10: # 단어 최대 길이 설정
        kor_titles.append(i)
    elif is_it_kor(i) == False and len(i) <= 10:
        eng_titles.append(i.lower())

kor_contents = []
eng_contents = []

for i in contents_data: 
    if is_it_kor(i) == True and len(i) <= 10:
        kor_contents.append(i)
    elif is_it_kor(i) == False and len(i) <= 10:
        eng_contents.append(i.lower())
```

Stopwords 생성 및 제거

```python
stopwords_string = 'this that to in' # 파이썬 질문게시판은 기본적으로 한국어기 때문에 많은 stopwords가 필요하지 않다고 판단했다.
eng_stopwords = stopwords_string.split()

eng_titles_1 = rem_trash(rem_stopwords(eng_titles, eng_stopwords))
eng_contents_1 = rem_trash(rem_stopwords(eng_contents, eng_stopwords))
```

```python
stopwords_string = '드립니다 입니다 겠습니다 있습니다 됩니다 습니다 하여도 합니다 하세요 하는데 에서 만약 해서 하고 하게 하면 같은 처럼 으로 요 을 에 때 하는 는 하는 를 로 과 가 이 의'
kor_stopwords = stopwords_string.split()

kor_titles_1 = rem_trash(rem_stopwords(kor_titles, kor_stopwords))
kor_contents_1 = rem_trash(rem_stopwords(kor_contents, kor_stopwords))



temp_data = {'data' : kor_titles_1+kor_contents_1+eng_titles+eng_contents_1}
data_df = pd.DataFrame(temp_data)


data_df.to_csv('data.csv', index=False, encoding='utf-8-sig')
```

워드클라우드 구현

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def wordcloud_maker(words, filename):
    text = ' '.join(words)
    wordcloud = WordCloud(background_color='white', font_path='AppleSDGothicNeoB.ttf', width=600, height=600, collocations=False).generate(text)
    plt.imshow(wordcloud, interpolation='bilInear')
    plt.axis('off')
    plt.show()
    wordcloud.to_file(filename + '.png')
```

```python
print('한글 제목 워드클라우드')
wordcloud_maker(kor_titles_1, 'kor_titles')
print('영어 제목 워드클라우드')
wordcloud_maker(eng_titles_1, 'eng_titles')
print('한글 내용 워드클라우드')
wordcloud_maker(kor_contents_1, 'kor_contents')
print('영어 내용 워드클라우드')
wordcloud_maker(eng_contents_1, 'eng_contents')
```

```python
all_of_words = kor_titles_1 + kor_contents_1 + eng_titles_1 + eng_contents_1
wordcloud_maker(all_of_words, 'all_of_words')
```
